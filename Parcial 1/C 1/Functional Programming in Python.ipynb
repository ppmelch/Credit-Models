# Functional Programming in Python

### Factorial Programming 

```python
def factorial(n):
    if n == 0 or n == 1: 
        return 1
    else:
        return n * factorial(n-1) 

factorial(4)
```

### Permutation Programming 

```python
def permutation(n, r):
    return factorial(n) / factorial(n-r)

permutation(4,2)
```

### Combinations Programming

```python
def combination(n, r):
    return permutation(n,r) * (1/factorial(r))

combination(4,2)
```

### Guess the number Game Rules

##### - The program generates a random number within a specified range (1-100)
##### - The player has a limited number of attempts to guess the number (5) 
##### - After each guess , the program provides feedback : too high , too low , or correct

```python
import random

def random_n():
    return random.randint(1, 100)

def adivina():
    r = random_n()
    max_attempts = 5
    attempts = 0

    while attempts < max_attempts:
        number = int(input("Ingresa el número: "))

        if number == r:
            print("Correct")
            return
        elif number < r:
            attempts += 1
            print(f"Too low — llevas {attempts} intentos, te quedan {max_attempts - attempts}")
        else:
            attempts += 1
            print(f"Too high — llevas {attempts} intentos, te quedan {max_attempts - attempts}")

    print(" Te quedaste sin intentos. El número era:", r)

adivina()

        




```

```python
class WeatherData:
    def __init__(self, temp_c, humidity):
        self.temp_c = temp_c
        self.humidity = humidity
    def display(self):
        print(f"Temperature: {self.temp_c}°C, Humidity: {self.humidity}%")
today = WeatherData(22, 45)
today.temp_c = 25
today.display()
```
