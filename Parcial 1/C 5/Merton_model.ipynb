```python
import numpy as np
from scipy.stats import norm
```

```python
V = 100_000_000
D = 60_000_000
rf = 0.03
vol = 0.2
T = 2
```

### Distance to Default 

```python
DD = (np.log(V/D) + (rf + (vol**2/2)) * T) / (vol * np.sqrt(T))
print(DD)
```

### Probability to Default (PD)

```python
print(1-norm.cdf(DD))
```

If the Debt change to 150 M

```python
D = 150_000_000
DD = (np.log(V/D) + (rf + (vol**2/2)) * T) / (vol * np.sqrt(T))
result = (1-norm.cdf(DD))*100
print(DD)
print(f'{result:.4f}%')
```
